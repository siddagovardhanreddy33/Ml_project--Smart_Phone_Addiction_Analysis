# ML Problem Statement

#### Smartphone addiction is increasing and affecting:

- sleep quality
- mental health
- academic/work performance
- stress levels
- We want to identify users who are likely addicted to smartphones using behavioural data.


#### Goal:
- Build a Machine Learning model to predict whether a person is addicted to smartphone or not using usage behaviour and lifestyle features.

## 2 Feature Engineering

```python
df
```

```python
# addictive behaviour hours
df["addictive_usage_hours"] = (df["social_media_hours"] + df["gaming_hours"])
df['addictive_usage_hours']
```

```python
df[['addictive_usage_hours','daily_screen_time_hours' ,'social_media_hours' ,'gaming_hours' ,'work_study_hours' ]]
```

```python
# addiction intensity ratio
df["addiction_usage_ratio"] = (df["addictive_usage_hours"]/ df["daily_screen_time_hours"])
df['addiction_usage_ratio']
```

```python
df[['addiction_usage_ratio', 'addictive_usage_hours','daily_screen_time_hours']]
```

```python

# fix unrealistic values
df["addiction_usage_ratio"] = df["addiction_usage_ratio"].clip(0,1)

df[['addiction_usage_ratio', 'addictive_usage_hours','daily_screen_time_hours']]
```

```python
df["addiction_usage_ratio"].describe()
```

```python
# sleep impact
df["sleep_deficit"] = 8 - df["sleep_hours"]
df[['sleep_deficit' , 'sleep_hours']]
```

```python

# weekend behaviour change
df["weekend_usage_diff"] = (df["weekend_screen_time"]- df["daily_screen_time_hours"])

df[['weekend_usage_diff','weekend_screen_time','daily_screen_time_hours']]
```

## 3 Train Test Split

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("addicted_label", axis=1)
y = df["addicted_label"]
```
